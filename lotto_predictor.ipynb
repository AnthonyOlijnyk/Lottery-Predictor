```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Dropout
```

```python
def calculate_minutes(date):
    total = 0
    mult = [43800, 1440, 525600]
    iso_date = date.split('/')
    for i in range(len(mult)):
        total = total + int(mult[i]) * int(iso_date[i])
    return total
```

```python
data = pd.read_csv("649.csv")
data.head()
```

```python
temp = pd.read_csv('toronto.csv')
temp.head()
```

```python
updated_date = []
trial = temp['date_time']
for element in trial:
    cart = element.split('-')
    for i in range(len(cart)):
        if cart[i].startswith('0'):
            cart[i] = cart[i].replace('0', '')
    phrase = cart[1] + '/' + cart[2] + '/' + cart[0]
    updated_date.append(calculate_minutes(phrase))
```

```python
temp['FIXED DATES'] = updated_date
temp.tail()
```

```python
dates = data["DRAW DATE"]
dates
```

```python
fixed_dates = []
for date in dates:
    fixed_dates.append(calculate_minutes(date))
len(fixed_dates)
```

```python
data['FIXED DATES'] = fixed_dates
data.head()
```

```python
data['maxtempC'] = None
data.head()
```

```python
df = pd.merge(data, temp, on='FIXED DATES')
len(df)
```

```python
features = df[['FIXED DATES', 'FeelsLikeC', 'sunHour', 'humidity', 'precipMM', 'windspeedKmph', 'WindGustKmph']]
labels = df[['NUMBER DRAWN 1', 'NUMBER DRAWN 2', 'NUMBER DRAWN 3', 'NUMBER DRAWN 4', 'NUMBER DRAWN 5', 'NUMBER DRAWN 6', 'BONUS NUMBER']]

X_train, X_test, y_train, y_test = train_test_split(features.to_numpy(), labels.to_numpy())

scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
model = Sequential()
model.add(Dense(7, input_shape=(7,), activation='linear'))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1028, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(7, activation='linear'))
```

```python
model.summary()
```

```python
model.compile(loss='mean_absolute_error', optimizer='rmsprop', metrics=['binary_accuracy'])
```

```python
model.fit(X_train_scaled, y_train, epochs=250, batch_size=64)
```

```python
print(model.predict(X_test_scaled[30:40]))
print(y_test[30:40])
print(X_test_scaled)
```
